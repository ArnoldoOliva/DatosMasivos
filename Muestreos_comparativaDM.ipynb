A primera vista el muestreo por estratos se asemeja mas, por lo que usaremos métodos multivariados para hacerlo más determinístico.

```python
#!pip install pingouin
import pingouin as pg
import pingouin as pg
from joblib import Parallel, delayed

# Datos de ejemplo
grupo1 = np.array(b[numerical])
grupo1 = np.asarray(grupo1, dtype=np.float64)

# Función para realizar la prueba de normalidad multivariada de Mardia en paralelo
def realizar_prueba(datos):
    resultado = pg.multivariate_normality(datos, alpha=0.05)
    return resultado

# Paralelizar el cálculo en múltiples núcleos/threads
num_nucleos = -1  # Utilizar todos los núcleos disponibles
resultados_paralelos = Parallel(n_jobs=num_nucleos)(delayed(realizar_prueba)(grupo1) for _ in range(100))

# Imprimir los resultados
for resultado in resultados_paralelos:
    print(resultado)
```

```python
# Deterministico: ESTRATIFICADO VS POBLACION

#ASUMIMOS QUE SON MULTIVARIADAS

# T de Hoteling multivariado:

import pingouin as pg

# Datos de ejemplo
grupo1 = estratificado[numerical].values
grupo2 = b[numerical].values
grupo1 = np.asarray(grupo1, dtype=np.float64)
grupo2 = np.asarray(grupo2, dtype=np.float64)

# Realizar la prueba t de Student multivariada
result = pg.multivariate_ttest(grupo1, grupo2, paired=False)

# Imprimir los resultados
print(result)

#H0=Medias multivariadas son iguales entre muestra y poblacion
#H0=Medias multivariadas NO son iguales entre muestra y poblacion


#pvalue mayor a 0.05, no RECHAZAMOS HO, por lo tanto, Medias multivariadas son iguales entre muestra y poblacion
```

```python
# Deterministico: CUOTA VS POBLACION

#ASUMIMOS QUE SON MULTIVARIADAS

# T de Hoteling multivariado:

# Datos de ejemplo
grupo1 = cuota[numerical].values
grupo2 = b[numerical].values
grupo1 = np.asarray(grupo1, dtype=np.float64)
grupo2 = np.asarray(grupo2, dtype=np.float64)

# Realizar la prueba t de Student multivariada
result = pg.multivariate_ttest(grupo1, grupo2, paired=False)

# Imprimir los resultados
print(result)

#H0=Medias multivariadas son iguales entre muestra y poblacion
#H0=Medias multivariadas NO son iguales entre muestra y poblacion


#pvalue menor a 0.05,  RECHAZAMOS HO, por lo tanto, Medias multivariadas NO son iguales entre muestra y poblacion
```

Usando T de Hotteling para comparar las medias de las diferentes variables a nivel multivariado entre las diferentes muestras (estratificado, donde se trabajaron con los cuantiles de numero de followers vs cuota, donde solo se selecciono los que no eran bots), vemos que en el caso del muestreo estratificado las medias muestrales son iguales a las poblacionales al no verse rechazado nuestra H0 de que los vectores de medias son iguales entre los dos set de datos, mientras que para el caso de la cuota si lo fueron, por lo tanto el muestreo estratificado fue mejor representando el set de datos.
